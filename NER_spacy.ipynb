## Imports

```python
import spacy
from spacy import displacy
```

## Spacy

```python
nlp = spacy.load('de_core_news_sm')

sen = "Schülerinnen und Schüler aus Schleswig-Holstein sind heute beim Tag der Wirtschaftsinformatik an der FH Kiel gemeinsam mit der CAU Kiel und der Dualen Hochschule Schleswig-Holstein."

text = nlp(sen)
displacy.render(text, style = 'ent', jupyter=True)
```

```python
nlp = spacy.load('de_core_news_sm')

sen = "Studierende nehmen heute an der Ringvorlesung der Dualen Hochschule Schleswig-Holstein in Kiel teil. Thema der heutigen Veranstaltung ist Named Entity Recognition (NLP)."

text = nlp(sen)
displacy.render(text, style = 'ent', jupyter=True)
```
